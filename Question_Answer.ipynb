```python
#Q1.9) Write a program to find the factors of a whole number using a while loop.
def find_factors(number):
    factors = []
    i = 1
    while i <= number:
        if number % i == 0:
            factors.append(i)
        i += 1
    return factors

# Example usage
num = int(input("Enter a whole number: "))
factors = find_factors(num)
print(f"The factors of {num} are: {factors}")
```

```python
#Q.1.11)Write a program to find prime numbers between 2 to 100 using nested loops.
def find_primes(start, end):
    primes = []
    for num in range(start, end + 1):
        is_prime = True
        for i in range(2, int(num**0.5) + 1):
            if num % i == 0:
                is_prime = False
                break
        if is_prime:
            primes.append(num)
    return primes

# Example usage
start = 2
end = 100
prime_numbers = find_primes(start, end)
print(f"Prime numbers between {start} and {end} are: {prime_numbers}")
```

```python
#8.Write a code for Fitness Club Manugernent System using 0-OPS:
#	create a Member class that has attributes such cis name, age, membership type,. and membership status (private).
# 	Implement methods to register a new member, renew a membership. and cancel a membership.
#	Use encapsulation to hide the member's unique identification number_
#	Inherit from the Member class to create a FamilyMernber class and an IndividualMember class, each with their own specific attributes and methods.

class Member:
    def __init__(self, member_id, name, age, membership_type):
        self.__member_id = member_id
        self.name = name
        self.age = age
        self.membership_type = membership_type
        self.__membership_status = 'Inactive'

    def register_member(self):
        self.__membership_status = 'Active'
        print(f"Member {self.name} (ID: {self.__member_id}) has been registered with a {self.membership_type} membership.")

    def renew_membership(self):
        if self.__membership_status == 'Active':
            print(f"Member {self.name} (ID: {self.__member_id}) has renewed their {self.membership_type} membership.")
        else:
            print(f"Member {self.name} (ID: {self.__member_id}) needs to register first.")

    def cancel_membership(self):
        if self.__membership_status == 'Active':
            self.__membership_status = 'Inactive'
            print(f"Member {self.name} (ID: {self.__member_id}) has canceled their membership.")
        else:
            print(f"Member {self.name} (ID: {self.__member_id}) is already inactive.")

    def get_membership_status(self):
        return self.__membership_status

    def get_member_id(self):
        return self.__member_id


class FamilyMember(Member):
    def __init__(self, member_id, name, age, family_members):
        super().__init__(member_id, name, age, 'Family')
        self.family_members = family_members

    def register_member(self):
        super().register_member()
        print(f"Family members: {', '.join(self.family_members)}")

    def renew_membership(self):
        super().renew_membership()
        print(f"Family members: {', '.join(self.family_members)}")

    def cancel_membership(self):
        super().cancel_membership()
        print(f"Family members: {', '.join(self.family_members)}")


class IndividualMember(Member):
    def __init__(self, member_id, name, age):
        super().__init__(member_id, name, age, 'Individual')


# Example usage
member1 = IndividualMember(1, 'John Doe', 30)
member2 = FamilyMember(2, 'Jane Smith', 40, ['Child 1', 'Child 2'])

# Register, renew, and cancel memberships
member1.register_member()
member1.renew_membership()
member1.cancel_membership()

member2.register_member()
member2.renew_membership()
member2.cancel_membership()
```

```python
#9.Write a code for Event management System using OOPS:
#•	Create an Event class chat has attributes such as name, date, time, location, and list of attendees (private).
#•	Implement methods to create a new event, add or remove attendees, and get the total number of attendees,
#•	Use encapsulation to hide the event's unique identification number,
#	Inherit from the Event class to create a FirivateEvent class and a PublicEvent class, each with their own specific attributes and methods.
class Event:
    def __init__(self, event_id, name, date, time, location):
        self.__event_id = event_id
        self.name = name
        self.date = date
        self.time = time
        self.location = location
        self.__attendees = []

    def create_event(self):
        print(f"Event '{self.name}' has been created on {self.date} at {self.time} in {self.location}.")

    def add_attendee(self, attendee):
        self.__attendees.append(attendee)
        print(f"Attendee '{attendee}' has been added to the event '{self.name}'.")

    def remove_attendee(self, attendee):
        if attendee in self.__attendees:
            self.__attendees.remove(attendee)
            print(f"Attendee '{attendee}' has been removed from the event '{self.name}'.")
        else:
            print(f"Attendee '{attendee}' is not in the list for the event '{self.name}'.")

    def get_total_attendees(self):
        return len(self.__attendees)

    def get_event_id(self):
        return self.__event_id

    def get_attendees(self):
        return self.__attendees


class PrivateEvent(Event):
    def __init__(self, event_id, name, date, time, location, host):
        super().__init__(event_id, name, date, time, location)
        self.host = host

    def create_event(self):
        super().create_event()
        print(f"This is a private event hosted by {self.host}.")


class PublicEvent(Event):
    def __init__(self, event_id, name, date, time, location, organizer):
        super().__init__(event_id, name, date, time, location)
        self.organizer = organizer

    def create_event(self):
        super().create_event()
        print(f"This is a public event organized by {self.organizer}.")


# Example usage
event1 = PrivateEvent(1, 'Birthday Party', '2024-08-15', '18:00', 'John\'s House', 'John Doe')
event2 = PublicEvent(2, 'Community Fair', '2024-09-10', '10:00', 'Central Park', 'City Council')

# Create events
event1.create_event()
event2.create_event()

# Add and remove attendees
event1.add_attendee('Alice')
event1.add_attendee('Bob')
event1.remove_attendee('Alice')

event2.add_attendee('Charlie')
event2.add_attendee('David')
event2.remove_attendee('Eve')

# Get total attendees
print(f"Total attendees for '{event1.name}': {event1.get_total_attendees()}")
print(f"Total attendees for '{event2.name}': {event2.get_total_attendees()}")
```

```python
#14.	Implement a Python module named string_utils.py containing functions for string manipulation, such as reversing and capitalizing strings.
# string_utils.py

def reverse_string(s):
    """Return the reversed version of the string s."""
    return s[::-1]

def capitalize_string(s):
    """Return the string s with the first character capitalized and the rest lowercased."""
    return s.capitalize()

def uppercase_string(s):
    """Return the string s with all characters in uppercase."""
    return s.upper()

def lowercase_string(s):
    """Return the string s with all characters in lowercase."""
    return s.lower()

def titlecase_string(s):
    """Return the string s with the first character of each word capitalized."""
    return s.title()

```

```python
#18.	Create a Python script that reads a text file named "expenses.txt" and calculates the total amount spent on various expenses listed in the file.
# Initialize the total amount spent
total_amount_spent = 0.0

try:
    # Open the file in read mode
    with open("expenses.txt", "r") as file:
        # Read and process each line
        for line in file:
            try:
                # Convert the line to a float and add it to the total
                amount = float(line.strip())
                total_amount_spent += amount
            except ValueError:
                print(f"Invalid line encountered and skipped: {line.strip()}")

    print(f"Total amount spent: ${total_amount_spent:.2f}")

except FileNotFoundError:
    print("The file expenses.txt does not exist.")
```

```python
#19.	Create a Python program that reads a text file named "poragraph,txt" and counts the occurrences of each ward in the paragraph, displaying the results in alphabetical order,
from collections import Counter
import re

# Initialize a Counter to store word occurrences
word_counts = Counter()

try:
    # Open the file in read mode
    with open("paragraph.txt", "r") as file:
        # Read the entire file content
        text = file.read()

        # Use regular expressions to find words and convert them to lowercase
        words = re.findall(r'\b\w+\b', text.lower())

        # Count the occurrences of each word
        word_counts.update(words)

    # Display the word counts in alphabetical order
    for word in sorted(word_counts):
        print(f"{word}: {word_counts[word]}")

except FileNotFoundError:
    print("The file paragraph.txt does not exist.")
```

```python
#21.	What do you mean by skewness.Explain its types.Ilse graph to show
"""Skewness
Definition:
Skewness is a measure of the asymmetry or deviation from the symmetry of a probability distribution. It indicates whether the data points in a dataset are concentrated on one side of the mean, leading to a longer tail on one side compared to the other. Skewness can be positive, negative, or zero.

Types of Skewness:

Positive Skewness (Right Skewed):
The tail on the right side (higher values) is longer or fatter than the left side.
The mean is greater than the median.
Example: Income distribution where a small number of people have very high incomes.

Negative Skewness (Left Skewed):
The tail on the left side (lower values) is longer or fatter than the right side.
The mean is less than the median.
Example: Age at retirement where most people retire around a certain age but a few retire much earlier.
Zero Skewness (Symmetrical):

The data is evenly distributed on both sides of the mean.
The mean, median, and mode are all equal.
Example: Heights of adult males in a large population.

Graphical Representation
Here’s a Python code to plot the different types of skewness"""

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate data for skewness examples
data_symmetrical = np.random.normal(loc=0, scale=1, size=1000)
data_positive_skew = np.random.exponential(scale=1, size=1000) - 1
data_negative_skew = -np.random.exponential(scale=1, size=1000) + 1

# Plot the data
fig, axs = plt.subplots(3, 1, figsize=(10, 15))

# Symmetrical distribution
sns.histplot(data_symmetrical, bins=30, kde=True, ax=axs[0])
axs[0].set_title('Symmetrical Distribution (Zero Skewness)')

# Positive skew distribution
sns.histplot(data_positive_skew, bins=30, kde=True, ax=axs[1])
axs[1].set_title('Positive Skewness (Right Skewed)')

# Negative skew distribution
sns.histplot(data_negative_skew, bins=30, kde=True, ax=axs[2])
axs[2].set_title('Negative Skewness (Left Skewed)')

plt.tight_layout()
plt.show()
```
